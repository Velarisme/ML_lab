```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
```

```python
url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
df = pd.read_csv(url)
df.head()
```

```python
print("Missing Values")
print(df.isnull().sum())
```

```python
df.loc[5:10, 'crim'] = np.nan
df.loc[15:18, 'age'] = np.nan

df['crim'] = pd.to_numeric(df['crim'], errors='coerce')
df.head()
```

```python
df['crim'] = df['crim'].fillna(df['crim'].mean())
df['age'] = df['age'].fillna(df['age'].median())
```

```python
Q1 = df['tax'].quantile(0.25)
Q3 = df['tax'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['tax'] < lower_bound) | (df['tax'] > upper_bound)]
print("Outliers:")
print(outliers)
print(outliers.shape[0])
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(y=df['tax'])
plt.title('Boxplot of Tax')
plt.xlabel('Tax Value')
plt.show()
```

```python
df['neighbours'] = np.random.choice(['s', 'n', 'k'], size=len(df))
lae = LabelEncoder()
df['neighbours_encoded'] = lae.fit_transform(df['neighbours'])
df.head()
```

```python
print(df.isnull().sum())
```

```python
mapping = {'s': 3, 'n': 6, 'k': 7}
df['neighbours_mapped'] = df['neighbours'].map(mapping)
df.head()
```
